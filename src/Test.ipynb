```python
import pandas as pd
import gzip
import json
```

```python
import os
import sys
from pathlib import Path
 
path = Path(os.path.dirname(os.getcwd()))
path = str(path)
print(path)
sys.path.insert(1, path)
```

```python
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')
```

```python
review_df = getDF(f'{path}/data/AMAZON_FASHION.json.gz')
review_df = review_df.rename({'overall':'rating'},axis = 1).copy()
review_df.head()
```
